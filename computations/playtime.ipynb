This notebook runs the reduced Khovanov-Rozansky computation for the positive braid `(s1, s2, s3, s1, s2, s3)` on the simply-laced Coxeter graph with triangle edges `12`, `23`, and `13`.

```python
from khovanov_rozansky import A, Q, T


def _as_polynomial(poly):
    return getattr(poly, "polynomial", poly)


def Euler_trace(poly):
    return _as_polynomial(poly).subs({T: -1})

def Point_count(poly):
    return _as_polynomial(poly).subs({T: -1, A: 0})

def Alexander_poly(poly):
    return _as_polynomial(poly).subs({T: -1, A: -Q**2})
```

```python
from khovanov_rozansky import DynkinDiagram, khovanov_rozansky_cohomology

diagram = DynkinDiagram.from_data([1, 2, 3], [(1, 2), (2, 3), (1, 3)])
braid = [(1, "+"), (2, "+"), (3, "+"), (1, "+"), (2, "+"), (3, "+")]

result = khovanov_rozansky_cohomology(diagram, braid, reduced=True)
result.polynomial
```

```python
Euler_trace(result)   # structural replacement, no simplification
```

```python
Alexander_poly(result)
```
